```python
import matplotlib.pyplot as plt
import pandas as pd

#Tesla combined ev sales and stock data
ev_stock_cleaned = pd.read_csv('EV_Data/data/cleaned_stock_data.csv')
ev_sales_cleaned = pd.read_csv('EV_Data/data/final_ev_sales.csv')
# Make sure 'Date' is datetime
ev_sales_cleaned['Date'] = pd.to_datetime(ev_sales_cleaned['Date'])
ev_stock_cleaned['Date'] = pd.to_datetime(ev_stock_cleaned['Date'])

# Merge sales and stock on Date for Tesla
tesla_data = pd.merge(ev_sales_cleaned[['Date', 'Tesla']], 
                      ev_stock_cleaned[['Date', 'Tesla_Close']], 
                      on='Date', how='inner')

fig, ax1 = plt.subplots(figsize=(12,6))

# Plot Tesla sales on left y-axis
color_sales = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Tesla EV Sales (units)', color=color_sales)
ax1.plot(tesla_data['Date'], tesla_data['Tesla'], color=color_sales, marker='o', label='Tesla Sales')
ax1.tick_params(axis='y', labelcolor=color_sales)

# Create second y-axis for stock price
ax2 = ax1.twinx()
color_stock = 'tab:red'
ax2.set_ylabel('Tesla Stock Closing Price', color=color_stock)
ax2.plot(tesla_data['Date'], tesla_data['Tesla_Close'], color=color_stock, linestyle='--', label='Tesla Stock Price')
ax2.tick_params(axis='y', labelcolor=color_stock)

plt.title('Tesla Monthly EV Sales vs. Stock Closing Price')
fig.tight_layout()
plt.savefig('Tesla_combined_chart.png', dpi=300)  # Save as PNG with high quality
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data once
ev_sales = pd.read_csv('EV_Data/data/final_ev_sales.csv')
ev_stock = pd.read_csv('EV_Data/data/cleaned_stock_data.csv')

# Convert 'Date' columns to datetime
ev_sales['Date'] = pd.to_datetime(ev_sales['Date'])
ev_stock['Date'] = pd.to_datetime(ev_stock['Date'])

# Define consistent colors for each company
colors = {
    'Tesla': 'tab:red',
    'BYD': 'tab:blue',
    'NIO': 'tab:green'
}

def plot_combined_sales_stock(company, sales_df, stock_df, save_path=None):
    """
    Plot combined monthly EV sales and stock closing price for a company.
    
    company: str, name like 'Tesla', 'BYD', or 'NIO'
    sales_df: DataFrame with 'Date' and sales columns
    stock_df: DataFrame with 'Date' and stock close price columns
    save_path: str or None, if given saves the plot to this filepath
    """
    # Merge on Date
    merged = pd.merge(
        sales_df[['Date', company]],
        stock_df[['Date', f'{company}_Close']],
        on='Date', how='inner'
    )

    fig, ax1 = plt.subplots(figsize=(12, 6))

    color = colors[company]

    # Sales line (left y-axis)
    ax1.plot(merged['Date'], merged[company], color=color, marker='o', label=f'{company} Sales')
    ax1.set_xlabel('Date')
    ax1.set_ylabel(f'{company} EV Sales (units)', color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    ax1.grid(True, which='both', axis='both', linestyle='--', alpha=0.3)

    # Stock price line (right y-axis)
    ax2 = ax1.twinx()
    ax2.plot(merged['Date'], merged[f'{company}_Close'], color=color, linestyle='--', label=f'{company} Stock Price')
    ax2.set_ylabel(f'{company} Stock Closing Price', color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    # Combine legends from both axes
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

    plt.title(f'{company} Monthly EV Sales vs. Stock Closing Price')
    fig.autofmt_xdate()
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300)

    plt.show()

# Example usage for all three companies
for c in ['Tesla', 'BYD', 'NIO']:
    plot_combined_sales_stock(c, ev_sales, ev_stock, save_path=f'{c}_combined_chart.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# 1. Load EV sales data
ev_sales = pd.read_csv('EV_Data/data/final_ev_sales.csv')
ev_sales['Date'] = pd.to_datetime(ev_sales['Date'])
ev_sales = ev_sales.sort_values('Date')

# 2. Download BYD stock data (ticker: 1211.HK)
byd_stock = yf.download('1211.HK', start='2024-01-01', end='2025-06-30', interval='1d')
byd_stock.reset_index(inplace=True)  # Date is index by default, so reset it
byd_stock['Date'] = pd.to_datetime(byd_stock['Date'])

# 3. Monthly average closing price for BYD
monthly_byd_stock = byd_stock.groupby(pd.Grouper(key='Date', freq='MS'))['Close'].mean().reset_index()

# 4. Download NIO stock data (ticker: NIO)
nio_stock = yf.download('NIO', start='2024-01-01', end='2025-06-30', interval='1d')
nio_stock.reset_index(inplace=True)
nio_stock['Date'] = pd.to_datetime(nio_stock['Date'])

# 5. Monthly average closing price for NIO
monthly_nio_stock = nio_stock.groupby(pd.Grouper(key='Date', freq='MS'))['Close'].mean().reset_index()

# 6. Plot BYD
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()

# Plot BYD Sales
ax1.plot(ev_sales['Date'], ev_sales['BYD'], label='BYD Sales', color='tab:blue', marker='o')
# Plot BYD Stock
ax2.plot(monthly_byd_stock['Date'], monthly_byd_stock['Close'], label='BYD Stock Price', color='tab:cyan', linestyle='--')

# Labels and titles
ax1.set_xlabel('Date')
ax1.set_ylabel('BYD Sales', color='tab:blue')
ax2.set_ylabel('BYD Stock Price (HKD)', color='tab:cyan')
plt.title('BYD: EV Sales vs. Stock Price')
fig.legend(loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.savefig('BYD_combined_chart.png', dpi=300)  # Save as PNG with high quality
plt.show()

# 7. Plot NIO
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()

# Plot NIO Sales
ax1.plot(ev_sales['Date'], ev_sales['NIO'], label='NIO Sales', color='tab:green', marker='o')
# Plot NIO Stock
ax2.plot(monthly_nio_stock['Date'], monthly_nio_stock['Close'], label='NIO Stock Price', color='tab:olive', linestyle='--')

# Labels and titles
ax1.set_xlabel('Date')
ax1.set_ylabel('NIO Sales', color='tab:green')
ax2.set_ylabel('NIO Stock Price (USD)', color='tab:olive')
plt.title('NIO: EV Sales vs. Stock Price')
fig.legend(loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.savefig('NIO_combined_chart.png', dpi=300)  # Save as PNG with high quality
plt.show()
```
